```python
# Function to remove outliers from a given feature in a dataset
def Outlier_remover(feature, feature_name, dataset):
    
    # Calculate the 25th and 75th quartiles
    Qrt_25, Qrt_75 = np.percentile(feature, 25), np.percentile(feature, 75)
    print('25th_Quartile: {} | 75th_Quartile: {}'.format(Qrt_25, Qrt_75))
    
    # Calculate the Interquartile Range (IQR)
    feat_iqr = Qrt_75 - Qrt_25
    print('IQR_:= {}'.format(feat_iqr))
    
    # Define the cutoff for identifying outliers
    FeatCutOff_ = feat_iqr * 1.5
    LowFeat, UppFeat = Qrt_25 - FeatCutOff_, Qrt_75 + FeatCutOff_
    print('Cut Off= {}'.format(FeatCutOff_))
    print(feature_name + ' Lower_:= {}'.format(LowFeat))
    print(feature_name + ' Upper_:= {}'.format(UppFeat))
    
    # Identify and count outliers
    outliers = [x for x in feature if x < LowFeat or x > UppFeat]
    print(feature_name + ' Outliers_of_cases_close_to_bankruptcy:= {}'.format(len(outliers)))
    
    # Remove outliers from the dataset
    dataset = dataset.drop(dataset[(dataset[feature_name] > UppFeat) | (dataset[feature_name] < LowFeat)].index)
    print('.' * 50)
    
    return dataset

# Apply the Outlier_remover function to all columns in the dataset
for col in Data_bank:
    new_dataframe = Outlier_remover(Data_bank[col], str(col), Data_bank)
```

```python
# here we are diving data & label

labels = new_dataframe['Bankrupt?']
new_dataframe = new_dataframe.drop(['Bankrupt?'], axis = 1)
```

```python
# Define a function for log transformation
def TransLog(data):
    for col in data:
        # Calculate the skewness of the column
        skew = data[col].skew()
        # Check if the skewness is significant (greater than 0.5 or less than -0.5)
        if skew > 0.5 or skew < -0.5:
            # Apply a log transformation to the column
            data[col] = np.log1p(data[col])
        else:
            # Skip columns with skewness within the acceptable range
            continue
    return data

# Apply the log transformation to the dataset and store the result in data_norm
data_norm = TransLog(new_dataframe)
```

```python
# Initialize lists to store evaluation metrics

###################LOGISTIC REGRESSION############################
AccuracyLstReg = []
PrecisionLstReg = []
RecallLstReg = []
F1LstReg = []
auc_lst_reg = []

# Create a Logistic Regression model
LogRegSm = LogisticRegression()

# Define hyperparameters for RandomizedSearchCV
LogRegParams = {"penalty": ['l2'],
                  'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                  'class_weight': ['balanced', None],
                  'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}

# Initialize RandomizedSearchCV with Logistic Regression and hyperparameters
RandLogReg = RandomizedSearchCV(LogisticRegression(), LogRegParams, n_iter=4)

# Perform stratified cross-validation
for train, val in sss.split(XTrainSM, YTrainSM):
    # Create a pipeline with SMOTE oversampling and Logistic Regression
    PipelineReg = imbalanced_make_pipeline(SMOTE(sampling_strategy='minority'), RandLogReg)
    ModelReg = PipelineReg.fit(XTrainSM[train], YTrainSM[train])  # Train the model
    BestEstReg = RandLogReg.best_estimator_  # Get the best estimator from RandomizedSearchCV
    RegPrediction = BestEstReg.predict(XTrainSM[val])  # Make predictions on the validation set
    
    # Store evaluation metrics for each fold
    AccuracyLstReg.append(PipelineReg.score(XTrainSM[val], YTrainSM[val]))
    PrecisionLstReg.append(precision_score(YTrainSM[val], RegPrediction))
    RecallLstReg.append(recall_score(YTrainSM[val], RegPrediction))
    F1LstReg.append(f1_score(YTrainSM[val], RegPrediction))
    auc_lst_reg.append(roc_auc_score(YTrainSM[val], RegPrediction))

# Display results
print('.' * 50)
print('-')
print('Logistic Regression (SMOTE) results:')
print('_')
print("Accuracy_:= {}".format(np.mean(AccuracyLstReg)))
print("Precision_:= {}".format(np.mean(PrecisionLstReg)))
print("Recall_:= {}".format(np.mean(RecallLstReg)))
print("F1_:= {}".format(np.mean(F1LstReg)))
print('_')
print('.' * 50)
```

```python
# Importing necessary libraries
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

# Calculating ROC curves for different classifiers
LogFpr, LogTpr, ThresholdLog = roc_curve(ValSm_Y, PredRegSm)
RfcFpr, RfcTpr, ThresholdRfc = roc_curve(ValSm_Y, SmotePredictionRfc)
XgbFpr, XgbTpr, ThresholdXgb = roc_curve(ValSm_Y, SmotePredictionXgb)
SVM_fpr, SVM_tpr, SVM_thresold = roc_curve(ValSm_Y, smote_prediction_SVM)

# Function to graph ROC curves for multiple classifiers
def graph_roc_curve_multiple(LogFpr, LogTpr, RfcFpr, RfcTpr, XgbFpr, XgbTpr, SVM_fpr, SVM_tpr):
    # Setting up the plot
    plt.figure(figsize=(20, 8))
    plt.title('ROC Curve', fontsize=14)
    
    # Plotting ROC curves for each classifier along with their AUC scores
    plt.plot(LogFpr, LogTpr, label='Logistic_Regression_Classifier_Score:= {:.4f}'.format(roc_auc_score(ValSm_Y, PredRegSm)))
    plt.plot(XgbFpr, XgbTpr, label='XGBoost_Classifier_Score:= {:.4f}'.format(roc_auc_score(ValSm_Y, SmotePredictionXgb)))
    plt.plot(SVM_fpr, SVM_tpr, label='SVM_Classifier_Score:= {:.4f}'.format(roc_auc_score(ValSm_Y, smote_prediction_SVM)))
    plt.plot(RfcFpr, RfcTpr, label='Random_Forest_Classifier Score:= {:.4f}'.format(roc_auc_score(ValSm_Y, SmotePredictionRfc)))
    
    # Plotting the diagonal line representing random guessing
    plt.plot([0, 1], [0, 1], 'k--')
    
    # Setting plot axis limits
    plt.axis([-0.01, 1, 0, 1])
    
    # Setting plot labels and annotations
    plt.xlabel('False Positive Rate', fontsize=13)
    plt.ylabel('True Positive Rate', fontsize=13)
    plt.annotate('Minimum ROC Score of 50% \n (This is the minimum score to get)', xy=(0.5, 0.5), xytext=(0.6, 0.3),
                arrowprops=dict(facecolor='#6E726D', shrink=0.05),
                )
    
    # Adding legend to the plot
    plt.legend()

# Calling the function with the calculated ROC curves
graph_roc_curve_multiple(LogFpr, LogTpr, RfcFpr, RfcTpr, XgbFpr, XgbTpr, SVM_fpr, SVM_tpr)

# Displaying the plot
plt.show()
```
